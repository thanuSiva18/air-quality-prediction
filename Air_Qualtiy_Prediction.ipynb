```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

# NOTE: You MUST run the following line in a Google Colab cell before running this code block:
# !pip install prophet
from prophet import Prophet

# ==============================================================================
# STEP 4: FEATURE ENGINEERING (Data Preparation for Prophet)
# ==============================================================================

# Prophet requires the datetime column to be named 'ds' and the target variable to be named 'y'.

TARGET_POLLUTANT = 'CO(GT)'

# 1. Select the target column and reset the index to turn 'datetime' into a column
prophet_df = df[[TARGET_POLLUTANT]].copy()
prophet_df.reset_index(inplace=True)

# 2. Rename the columns to Prophet's required format
prophet_df.rename(columns={'datetime': 'ds', TARGET_POLLUTANT: 'y'}, inplace=True)

print("Prophet DataFrame Ready:")
print(prophet_df.head())

# ==============================================================================
# STEP 5: MODEL DEVELOPMENT
# ==============================================================================

# 1. Split the data into training and testing sets
# We split chronologically (80% for training, 20% for testing)
train_size = int(len(prophet_df) * 0.8)
train = prophet_df[:train_size]
test = prophet_df[train_size:]

print(f"\nTraining Data Length: {len(train)}")
print(f"Testing Data Length: {len(test)}")

# 2. Initialize and Train the Prophet Model
# NOTE: We can add seasonality_mode='multiplicative' for air quality data
model = Prophet(
    daily_seasonality=True,
    weekly_seasonality=True,
    yearly_seasonality=True,
    seasonality_mode='multiplicative'
)

# Train the model
model.fit(train)

# 3. Predict Future Air Quality Levels
# Create a future DataFrame to make predictions for the testing period
future_dates = model.make_future_dataframe(periods=len(test), include_history=False)

# Generate the forecast
forecast = model.predict(future_dates)

# Combine the forecast with the actual test data for comparison
y_true = test['y'].values
y_pred = forecast['yhat'].values

# ==============================================================================
# STEP 6: MODEL EVALUATION AND VISUALIZATION
# ==============================================================================

print("\n--- Model Evaluation Metrics ---")

# 1. Calculate Evaluation Metrics
mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
r2 = r2_score(y_true, y_pred)

print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R-squared (R2): {r2:.4f}")

# 2. Visualize Actual vs. Predicted Pollutant Levels
plt.figure(figsize=(14, 7))
plt.plot(test['ds'], y_true, label='Actual Values', color='blue', linewidth=2)
plt.plot(future_dates['ds'], y_pred, label='Predicted Forecast', color='red', linestyle='--', linewidth=2)
plt.fill_between(
    future_dates['ds'],
    forecast['yhat_lower'],
    forecast['yhat_upper'],
    color='red',
    alpha=0.2,
    label='Prediction Interval (80%)'
)

plt.title(f'Actual vs. Predicted {TARGET_POLLUTANT} Concentration (Test Set)')
plt.xlabel('Date and Time')
plt.ylabel(f'{TARGET_POLLUTANT} Concentration')
plt.legend()
plt.grid(True, alpha=0.5)
plt.xticks(rotation=45)
plt.tight_layout()

# Save the plot
plt.savefig('actual_vs_predicted_air_quality.png')
plt.close()

print("\nVisualization saved as 'actual_vs_predicted_air_quality.png'")
```
